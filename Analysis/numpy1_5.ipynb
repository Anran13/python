```python
import numpy as np
import pandas as pd

df = pd.read_csv('world.csv', index_col='國家')
print(df.head(5))
print('==============================================================\n')

condition = df.index == "台灣"
taiwan = df[condition]
print(taiwan[['日期','總確診數']].tail(5))
print('==============================================================\n')

print(taiwan[taiwan['新增確診數'] > 100])

print('==============================================================\n')

# save the dataframe
taiwan.to_csv('taiwan.csv',encoding='utf-8',index=True)
```
