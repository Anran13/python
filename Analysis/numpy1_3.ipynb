```python
# normal distribution

import numpy as np
import matplotlib.pyplot as plt

# print(np.random.normal(loc=0, scale=1, size=10))
# print(np.random.normal(loc=0, scale=1, size=(2,5)))
n = 1000
normal_dist = np.random.normal(loc=0, scale=1, size=n)
plt.hist(normal_dist, bins=30, color='#EB7A77')
# colors
# https://nipponcolors.com/
plt.show()
```

```python
from scipy.stats import norm
# In terminal(root filepath): enter "pip install -r requirements.txt"

plt.hist(normal_dist, bins=30, color='#EB7A77')
# find the minimum and maximum value of x axis according to the previous plot
xmin, xmax = plt.xlim()
print((xmin, xmax))

x = np.linspace(xmin, xmax, n)
# print(x)
# print(np.mean(normal_dist))
# print(np.std(normal_dist))
prob = norm.pdf(x, loc=np.mean(normal_dist), scale=np.std(normal_dist))
# print(prob)
figure, axes = plt.subplots(constrained_layout = True, figsize=(4,3))
axes.hist(normal_dist, bins=30)
axes.set_ylabel("frequency")
axes.set_xlim((min(x), max(x)))
# Dual-axis setting
axes2 = axes.twinx()
axes2.plot(x, prob*100, color='red')
axes2.set_ylabel("probability (%)")
plt.show()
```
