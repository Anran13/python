```python
import math
import numpy as np

x = np.arange(0, 2.5, 0.1)
# x:ndarray = np.arange(0, 2.5, 0.1)
import matplotlib.pyplot as plt

print(x)
print(x.ndim)
print(x.size)
print(x.shape)
print(x.dtype)

y1 = np.sin(x * math.pi)
y2 = np.sin(x * math.pi + math.pi/2)

# using module plt to plot
figure, axes = plt.subplots(constrained_layout = True)
plt.plot(x, y1,marker='*',label="90")
plt.plot(x, y2,marker='s',label="0")
plt.legend(loc='lower right')
# axes.legend(loc='upper right', bbox_to_anchor=(1.18, 1))
plt.title("sin()")
plt.xlabel("pi")
plt.ylabel("sin")
plt.show()
```

```python
import math
import numpy as np
# from numpy import ndarray

x = np.arange(0, 2.5, 0.1)
# x:ndarray = np.arange(0, 2.5, 0.1)
import matplotlib.pyplot as plt

print(x)
print(x.ndim)
print(x.size)
print(x.shape)
print(x.dtype)

y1 = np.sin(x * math.pi)
y2 = np.sin(x * math.pi + math.pi/2)

# using instance under module plt to plot
figure, axes = plt.subplots(constrained_layout = True)
axes.plot(x, y1,marker='*',label="90")
axes.plot(x, y2,marker='s',label="0")
axes.legend(loc='lower right')
# axes.legend(loc='upper right', bbox_to_anchor=(1.18, 1))
axes.set_title("sin()")
axes.set_xlabel("pi")
axes.set_ylabel("sin")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors =['gray', 'green', 'red', 'blue']
explode = (0, 0, 0.1, 0)

plt.pie(values, labels=labels, colors=colors, explode=explode, startangle=-65, shadow=True)
```

```python
_, _, autotexts = plt.pie(values,labels=labels,colors=colors,explode=explode,shadow=True,startangle=-65,autopct='%1.1f%%')
plt.setp(autotexts, size=14, color='white')
plt.show()
```
