```python
import math
import numpy as np
# from numpy import ndarray

x = np.arange(0, 2.5, 0.1)
# x:ndarray = np.arange(0, 2.5, 0.1)
import matplotlib.pyplot as plt

print(x)
print(x.ndim)
print(x.size)
print(x.shape)
print(x.dtype)

y1 = np.sin(x * math.pi)
y2 = np.sin(x * math.pi + math.pi/2)

figure, axes = plt.subplots()
axes.plot(x, y1,marker='*',linestyle='',label="90")
axes.plot(x, y2,marker='s',linestyle='',label="0")
axes.legend(loc='upper right', bbox_to_anchor=(1.18, 1))
axes.set_title("sin()的正弦波")
axes.set_xlabel("圓周率")
axes.set_ylabel("正弦波值")
plt.show()
```
