```python
scores = list(range(10, 60, 10))
new_scores = list()
print(len(scores))

for i in scores:
    i += 5
    new_scores.append(i)

print(new_scores)
```

```python
new_scores = [i+5 for i in scores]
print(new_scores)
```

```python
import random

students = np.array([[60, 82, 85, 90, 89],
                    [56, 77, 78, 80, 75]])
print(students)

print(students[0]) 
print(students[1])
print(students[1,:3])
print(students[0,-3:])
print(students.shape)
print(len(students))
```

```python
subjects = ['chinese', 'english', 'math', 'history', 'chemistry']
print(subjects)
```

```python
# plotting
import matplotlib.pyplot as plt

for id in list(range(len(students))):
    print(students[id])

print([students[id] for id in list(range(len(students)))])

figure = plt.figure()
axes = figure.add_subplot()
# print(figure)

axes.plot(subjects, students[0], label = 'student 1')
axes.plot(subjects, students[1], label = 'student 2')
axes.legend()
axes.set_title("student 1 vs student 2")
axes.set_xlabel('subjects')
axes.set_ylabel('scores')
print(figure)

plt.savefig("plot1.png")
plt.show
plt.close
```

```python
# subplot

figure = plt.figure(figsize=(10,6))
axes1 = figure.add_subplot(1,2,1)
axes2 = figure.add_subplot(1,2,2)
# print(figure)

axes1.plot(subjects, students[0], label = 'student 1')
axes2.plot(subjects, students[1], label = 'student 2')
axes1.legend()
axes2.legend()
axes1.set_xlabel('subjects')
axes1.set_ylabel('scores')
axes2.set_xlabel('subjects')
axes2.set_ylabel('scores')
print(figure)

plt.savefig("plot2.png")
plt.show
plt.close
```
